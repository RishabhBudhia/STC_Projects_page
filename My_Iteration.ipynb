**2. DIVIDING DATA INTO TRAINING SET AND TESTING SET**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score ,mean_squared_error ,mean_absolute_error
```

```python
y=df["ANNUAL"] #the column that needs to be predicted
X=df[["Jan-Feb","Mar-May","Jun-Sep","Oct-Dec"]] #Input data to predict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #splitting the data for model training
```

###  **Linear Regression**

```python
model=LinearRegression() #creating an instance of the LinearRegression() model
```

```python
model.fit(X_train,y_train) #Training the data
```

```python
predsLR=model.predict(X_test) #predicting the values from the test data and storing in preds
```

```python
predsLR #this is how predictions look like
```

### **RANDOM FOREST Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predsRF=model.predict(X_test)
```

### **SVM Regressor**

```python
from sklearn.svm import SVR
```

```python
model =SVR()
```

```python
model.fit(X_train,y_train)
```

```python
predsSVM = model.predict (X_test)
```

### **K Nearest Neigbors**

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model= KNeighborsRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
predsKNN =model.predict(X_test)
```

## **Comparision of Models**

```python
# Actual Value : y_test

#Predicted values by different models:

predictions ={"Linear Regression":predsLR ,"Random Forest Regressor" : predsRF ,"SVM Regressor" : predsSVM ,"KNN" :predsKNN}
```

```python
print("Root Mean Squared Error :")
for k,v in predictions.items():
 print("Model {} : {} ".format(k,mean_squared_error(y_test,v,squared =False)))
```

```python
print("R2 Score :")
for k,v in predictions.items():
 print("Model {} : {} ".format(k,r2_score(y_test,v,)))
```

```python
plt.figure(figsize=(15,15))
plt.plot(y_test - predsLR,linestyle=' ',marker="o",label="Linear Regresion")
plt.plot(y_test - predsRF,linestyle=' ',marker="x",label="Random Forest")
plt.plot(y_test -predsSVM,linestyle=' ',marker="X",label="SVM")
plt.plot(y_test -predsKNN,linestyle=" ",marker = "+",label="KNN")
plt.title("Plotting Errors",fontweight="bold",fontsize=15)
plt.legend()

#Error must be centered around 0 
#SVM Deviates alot 
#Random Forest has some deiviation
#Linnear Regression and KNN Performs Well
```

```python
plt.figure(figsize=(10,10))
sns.regplot(x=y_test ,y= predsLR,label="Linear Regression")
sns.regplot(x=y_test,y=predsRF,label="Random Forest")
sns.regplot(x=y_test ,y=predsSVM,marker="X",label="SVM")
sns.regplot(x=y_test ,y=predsKNN,marker = "+",label="KNN")
plt.title("Line Plot to Check Variance",fontweight="bold",fontsize=15)
plt.legend()
```
